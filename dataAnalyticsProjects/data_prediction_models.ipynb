##### Analyzing for outliers 

```python
import matplotlib.pyplot as plt
import seaborn as sns

df_cp = df.copy()
plt.figure(figsize=(25,25))

df_cp.boxplot(column=['MinTemp', 'MaxTemp', 'Rainfall', 'Evaporation', 'Sunshine',
                   'WindGustSpeed', 'WindSpeed9am', 'WindSpeed3pm', 
                   'Humidity9am', 'Humidity3pm', 'Pressure9am', 'Pressure3pm', 
                   'Cloud9am', 'Cloud3pm', 'Temp9am', 'Temp3pm'])

plt.xticks(rotation=45, fontsize=12)

plt.grid(True)

plt.title('Boxplots of Weather Features', fontsize=16)
plt.ylabel('Value Range', fontsize=12)
sns.boxplot(df_cp)
```

```python
# Sum of each attributes outliers 
def detect_outliers_iqr(data):
    Q1 = data.quantile(0.25)
    Q3 = data.quantile(0.75)
    IQR = Q3 - Q1
    outliers = (data < (Q1 - 1.5 * IQR)) | (data > (Q3 + 1.5 * IQR))
    return outliers

numerical_cols = ['MinTemp', 'MaxTemp', 'Rainfall', 'Evaporation', 'Sunshine',
                   'WindGustSpeed', 'WindSpeed9am', 'WindSpeed3pm', 
                   'Humidity9am', 'Humidity3pm', 'Pressure9am', 'Pressure3pm', 
                   'Cloud9am', 'Cloud3pm', 'Temp9am', 'Temp3pm']
for col in numerical_cols:
    outliers = detect_outliers_iqr(df_cp[col])
    print(f"Outliers in {col}:")
    print(outliers[outliers].sum())
```

```python
numerical_cols = ['MinTemp', 'MaxTemp', 'Rainfall', 'Evaporation', 'Sunshine',
                   'WindGustSpeed', 'WindSpeed9am', 'WindSpeed3pm', 
                   'Humidity9am', 'Humidity3pm', 'Pressure9am', 'Pressure3pm', 
                   'Cloud9am', 'Cloud3pm', 'Temp9am', 'Temp3pm']

for column in numerical_cols:
    column_vals = df_cp[column]

    Q1 = column_vals.quantile(0.25)
    Q3 = column_vals.quantile(0.75)
    IQR = Q3-Q1

    lowbound = Q1 - 1.5*IQR
    upbound = Q3 + 1.5*IQR

    median = column_vals.median()
    df_cp[column] = df_cp[column].apply(lambda x: median if(x<lowbound or x>upbound) else x)

plt.figure(figsize=(25, 25))
sns.boxplot(data=df_cp[numerical_cols])
plt.show()
```

#### Imputing the catogorical values with mode.

```python
df_cp['WindDir9am'].fillna(df_cp['WindDir9am'].mode()[0], inplace=True)

df_cp['WindDir3pm'].fillna(df_cp['WindDir3pm'].mode()[0], inplace=True)

df_cp['WindGustDir'].fillna(df_cp['WindGustDir'].mode()[0], inplace= True)

print(df_cp.isna().sum())
print(df_cp.isna().sum().sum())
```

#### Imputing the missing values with the mean and median according to the data.

```python
# Impute the missing values with the mean for which data has a bell curve
mean_cols = ['MinTemp', 'MaxTemp', 'Humidity9am', 'Humidity3pm', 
             'Pressure9am', 'Pressure3pm', 'Temp9am','Temp3pm']

# Impute the missing values with the median for which data is skewed 
median_cols = ['Rainfall', 'Evaporation', 'Sunshine', 'WindGustSpeed', 
               'WindSpeed9am', 'WindSpeed3pm', 'Cloud9am', 'Cloud3pm']

for col in mean_cols:
    df_cp[col].fillna(df_cp[col].mean(), inplace=True)

for col in median_cols:
    df_cp[col].fillna(df_cp[col].median(), inplace=True)

print(df_cp.isna().sum())
print(df_cp.isna().sum().sum())
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
import numpy as np
import matplotlib.pyplot as plt

#Determining the optimal value of estimators for the model
# Initialization
auc_scores = []

# Class weights to balance the classes
weights = {0: 0.6449270797240684, 1: 2.225005433601391}

# Different numbers of estimators to evaluate
estimators = [10, 20, 30, 60, 120, 240, 500]

# Iterating over different values of n_estimators
for n in estimators:
    # Define the RandomForestClassifier
    r = RandomForestClassifier(n_estimators=n, n_jobs=-1, class_weight=weights)
    
    # Perform 5-fold cross-validation and calculate mean AUC score
    cv_auc_scores = cross_val_score(r, X_train, y_train, cv=5, scoring='roc_auc')
    auc_scores.append(np.mean(cv_auc_scores))

# Plotting the AUC scores against the number of estimators
plt.figure(figsize=(12, 8))
plt.plot(estimators, auc_scores, marker='o')
plt.title('AUC Score vs. Number of Estimators in Random Forest')
plt.xlabel('Number of Estimators')
plt.ylabel('Mean AUC Score')
plt.xticks(estimators)
plt.grid(True)
plt.show()
```

```python
# Function to detect outliers using IQR for the test set
def detect_outliers_iqr_test(data_test):
    Q1 = data_test.quantile(0.25)
    Q3 = data_test.quantile(0.75)
    IQR = Q3 - Q1
    outliers_test = (data_test < (Q1 - 1.5 * IQR)) | (data_test > (Q3 + 1.5 * IQR))
    return outliers_test

# List of numerical columns in the test set
numerical_cols_test = ['MinTemp', 'MaxTemp', 'Rainfall', 'Evaporation', 'Sunshine',
                       'WindGustSpeed', 'WindSpeed9am', 'WindSpeed3pm', 
                       'Humidity9am', 'Humidity3pm', 'Pressure9am', 'Pressure3pm', 
                       'Cloud9am', 'Cloud3pm', 'Temp9am', 'Temp3pm']

# Loop through each numerical column in the test set to detect and handle outliers
for column_test in numerical_cols_test:
    column_vals_test = df_test_cp[column_test]

    Q1_test = column_vals_test.quantile(0.25)
    Q3_test = column_vals_test.quantile(0.75)
    IQR_test = Q3_test - Q1_test

    lowbound_test = Q1_test - 1.5 * IQR_test
    upbound_test = Q3_test + 1.5 * IQR_test

    median_test = column_vals_test.median()
    df_test_cp[column_test] = df_test_cp[column_test].apply(lambda x: median_test if (x < lowbound_test or x > upbound_test) else x)

# Plotting boxplot to visualize the columns in the test set after handling outliers
plt.figure(figsize=(25, 25))
sns.boxplot(data=df_test_cp[numerical_cols_test])
plt.show()
```
